```python
#Advanced Indexing and Selection
# Create a list of the passengers' names and ages for those passengers under the age of 21 
passengers = data[data.Age < 21][['Name', 'Age']]
passengers
```

```python
#Advanced Indexing and Selection
#Count how many passengers were between the ages of 21 and 30. Do not use
#Python's and logical operator for this step, but rather the ampersand symbol (&).

adults = data.loc[ (data['Age'] > 21) & (data.Age < 30)]
len(adults)
```

```python
#Count the passengers that were either fist- or third-class ticket holders. Again,
#we will not use the Python logical or operator but rather the pipe symbol (| ). 
passenger_class = data.loc[ (data.Pclass == 1) | (data.Pclass == 3 )  ]

print("Count the passengers in Class I or Class III:", len(passenger_class))
```
